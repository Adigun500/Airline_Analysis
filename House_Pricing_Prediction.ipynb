### **Data Dictionary of House Pricing Dataset**

The data set contains 506 observations of house pricing from different towns. Corresponding to each house price, data of 18 other variables is available on which price is suspected to depend.


price:- Value of the house.

crime_rate:- Crime rate in that neighborhood.

resid_area:- proportion of residential area in the town.

air_qual:- Qualitty of air in that neighborhood.

room_num:- Average number of rooms in houses of that locality.

age:- How old is the house construction in years.

dist1:- Distance from employment hub 1.

dist2:- Distance from employment hub 2.

dist3:- Distance from employment hub 3.

dist4:- Distance from employment hub 4.

teachers:- Number of teachers per thousand population.

poor_prop:- Proportion of poor population in the town.

airport:- Is there an airport in the city? (Yes/No)

n_hos_beds:- Number of hospital beds per 1000 population in the town.

n_hot_rooms:- Number of hotel rooms per 1000 population in the town.

waterbody:- What type of natural fresh water source is there in the city (lake/river/both/none)

rainfall:- The yearly average rainfall in centimeters

bus_ter:- Is there a bus terminal in the city? (Yes/No)

parks:- Proportion of land assigned as parks and green areas in the town.

# **Treatment of Outliers**

```python
#identifying the 99 and 1 percentile value

uv=np.percentile(df.n_hot_rooms,[99])[0]

uv
```

```python
#We get data that are greater than uv

df[(df.n_hot_rooms>uv)]
```

```python
#Now we need to apply capping and flooring.

df.n_hot_rooms[(df.n_hot_rooms>3*uv)] = 3*uv
```

```python
#cofirming the result

df[(df.n_hot_rooms>uv)]
```

```python
#Treating the outlier of the rainfall variables, save the result to lv(lowervalue)

lv = np.percentile(df.rainfall,[1])[0]

lv
```

```python
#Getting the outlier data

df[(df.rainfall<lv)]
```

```python
#applying capping and flooring

df.rainfall[(df.rainfall <0.3*lv)] = 0.3*lv
```

```python
#confirming the result

df[(df.rainfall<lv)]
```

```python
#to confirm if crime rate has outliers

sns.jointplot(x="crime_rate", y='price',data=df)
```

From the above viz, it can be seen that, the lower the crime rate, the higer the price, and vice versa.

```python
#now let's check our data to see the transformation we have done so far

df.describe()
```

```python
# To do linear regression we need to import the statsmodels.api

import statsmodels.api as sn
```

```python
# Creating constant for our model, the first one here will be our betta not

X = sn.add_constant(df['room_num'])

```

```python
#For the model, we will create a model object

lm = sn.OLS(df['price'],  X).fit()
```

```python
#To get the summary of the model

lm.summary()
```

```python
#Trying out the second method of running regression analysis in python

from sklearn.linear_model import LinearRegression
```

```python
#For sklearn, we need to define our x and y variables. This is done as 1d nparray

y=df['price']
```

```python
#Declare x varible and it has to be 2d nparray.

x=df[['room_num']]
```

```python
#We need to create our linear regression object

lm2 = LinearRegression()
```

```python
#Then we can fit our model to lm2

lm2.fit(x,y)
```

```python
#Now to see the intercept and coefficient

print(lm2.intercept_,lm2.coef_)
```

From the above result, our betta not is -34 and our betta one(coefficient of x) is 9

```python
#To predict the value of y base on the linear model we generated

lm2.predict(x)
```

```python
#To plot our regression line

sns.jointplot(x=df['room_num'], y=df['price'],data=df,kind='reg')
```
